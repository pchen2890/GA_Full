```python
import numpy
from scipy.stats import mode

def five(key,salesdata):
# Print out the column name
    print key
# Print the mean of the data using np.mean()
    print mean(salesdata[key])
# Print out the median of the data using np.median()
    print median(salesdata[key])
# Print out the mode of the rounded data using stats.mode()
    dist_rounded = [int(round(n)) for n in salesdata[key]]
    print mode(dist_rounded)
# Print out the variance of the data using np.var()
    print var(salesdata[key])
# Print out the standard deviation of the data using np.std()
    print std(salesdata[key])

five("volume_sold",salesdata)
```

```python
import seaborn as sns
%matplotlib inline
def distribution_plotter(column, data):
    sns.set(rc={"figure.figsize": (10, 7)})
    sns.set_style("white")
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    dist.set_title('Distribution of ' + column + '\n', fontsize=16)
    
distribution_plotter("volume_sold",salesdata["volume_sold"])
```

```python
distribution_plotter("2015_q1_sales",salesdata["2015_q1_sales"])
```
